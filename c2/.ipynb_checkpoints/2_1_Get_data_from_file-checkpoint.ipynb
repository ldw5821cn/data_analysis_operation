```python
xlsx = xlrd.open_workbook('demo.xlsx')
print('All sheet: %s'%xlsx.sheet_names())
```

```python
sheet1 = xlsx.sheets()[0]
sheet1_name = sheet1.name
sheet1_cols = sheet1.ncols
sheet1_nrows = sheet1.nrows
print(('Sheet1 Name: %s\nSheet1 cols: %s\nSheet1 rows: %s') % (sheet1_name, sheet1_cols, sheet1_nrows))
```

```python
sheet1_nrows4 = sheet1.row_values(4)
sheet1_cols2 = sheet1.col_values(2)
cell123 = sheet1.row(2)[3].value
print(('Row 4: %s\nCol 2: %s\nCell 1: %s\n') % (sheet1_nrows4, sheet1_cols2, cell123))
```

```python
for i in range(sheet1_nrows):
    print(sheet1.row_values(i))
```

```python
add = '北京市东直门'
ak = '3XWEB0OkQADeGyyOPcvMZShtQCDH2eDd'
#http://api.map.baidu.com/geocoder
#?location=39.990912172420714,116.32715863448607&coord_type=gcj02&output=html&src=webapp.baidu.openAPIdemo
url = 'http://api.map.baidu.com/geocoder?address=%s&output=json&ak=%s'
res = requests.get(url % (add,ak))
add_info_json = res.text
print(add_info_json)
```

```python
add_json = json.loads(add_info_json)
lat_lng = add_json['result']['location']
print(lat_lng)
```

```python
add = '北京市中关村软件园'
ak = '3XWEB0OkQADeGyyOPcvMZShtQCDH2eDd'
#http://api.map.baidu.com/geocoder
#?location=39.990912172420714,116.32715863448607&coord_type=gcj02&output=html&src=webapp.baidu.openAPIdemo
url = 'http://api.map.baidu.com/geocoder?address=%s&output=xml&ak=%s'
res_xml = requests.get(url % (add,ak))
add_info_xml = res_xml.text
print(add_info_xml)
```

```python
root  = Etree.fromstring(add_info_xml)
lng = root[1][0][0].text
lat = root[1][0][1].text
print('lng: %s' % lng)
print('lat: %s' % lat)
```
