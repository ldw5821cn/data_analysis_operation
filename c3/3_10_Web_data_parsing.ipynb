```python
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
# 获取总页面数量
def get_total_page_number():
    user_agent = 'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101 Safari/537.36'
    headers = {'User-Agent': user_agent}  # 定义头信息
    url = 'https://www.amazon.cn/s/ref=sa_menu_digita_l3_siphone?ie=UTF8&page=1&rh=n%3A665002051%2Cp_89%3AApple%2Cn%3A664978051'  # 寻找页码的URL
    res = requests.get(url, headers=headers)  # 发送请求
    html = res.text  # 获得请求中的返回文本信息
    html_soup = BeautifulSoup(html,"lxml")  # 建立soup对象，用于处理HTML
    page_number_span = html_soup.find('span', id="s-result-count")  # 查找id="s-result-count"的h2标签
    page_number_code = page_number_span.text  # 读出该标签的文本信息
    print(page_number_code)
    number_list = re.findall('\d+', page_number_code)  # 使用正则表达式解析出文本中的3个数字
    print(number_list)
    total_page_number = (int(number_list[-1]) / int(number_list[-2])) + 1  # 计算得出总页码
    return int(total_page_number)  # 返回页面数字
```

```python
def parse_single_page(i):
    url_part1 = 'https://www.amazon.cn/s/ref=sa_menu_digita_13_siphone?ie=UTF8&page=%d' % i
    url_part2 = '&rh=n%3A665002051%2Cp_89%3AApple%2Cn%3A664978051'
    url = url_part1 + url_part2
    print('parse url:%s' % url)
    user_agent = 'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101 Safari/537.36'
    headers = {'User-Agent':user_agent}
    url = 'https://www.amazon.cn/s/ref=sa_menu_digita_l3_siphone?ie=UTF8&page=1&rh=n%3A665002051%2Cp_89%3AApple%2Cn%3A664978051'
    res = requests.get(url, headers=headers)
    html = res.text
    html_soup = BeautifulSoup(html,"lxml")
    tag_list = html_soup.find_all('li',id=re.compile('^result.*'))
    for tag_info in tag_list:
        print(tag_info)
        price_code = tag_info.find('span',class_="a-size-base a-color-price s-price a-text-bold")
        if price_code != None:
            price = price_code.text
            title_code = tag_info.find('h2')
            title = title_code.text
            write_data(title, price)
```

```python
def write_data(title,price):
    file_data = time.strftime('%Y-%m-%d',time.localtime(time.time()))
    fn = open('%s.txt' % file_data, 'a+')
    content = title + '\t' + price + '\n'
    fn.write(content)
    fn.close()
```

```python
def main():
    total_page_number = get_total_page_number()
    for i in range(1,int(total_page_number)+1):
        parse_single_page(i)
```

```python
main()
```
