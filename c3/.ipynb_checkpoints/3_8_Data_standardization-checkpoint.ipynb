```python
#python 数据标准化处理
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data6.txt', delimiter='\t')
zscore_scaler = preprocessing.StandardScaler()
data_scale_1 = zscore_scaler.fit_transform(data)
```

```python
minmax_scaler = preprocessing.MinMaxScaler()
data_scale_2 = minmax_scaler.fit_transform(data)
maxabsscaler_scaler = preprocessing.MaxAbsScaler()
data_scale_3 = maxabsscaler_scaler.fit_transform(data)
```

```python
robustscalerr_scaler = preprocessing.RobustScaler()
data_scale_4 = robustscalerr_scaler.fit_transform(data)
```

```python
data_list = [data,data_scale_1,data_scale_2,data_scale_3,data_scale_4]
scalar_list = [15,10,15,10,15,10]
color_list = ['black','green','blue','yellow','red']
merker_list = ['o',',','+','s','p']
title_list = ['source data','zscore_scaler','minmax_scaler','maxabsscaler_scaler','robustscalerr_scaler']
for i, data_single in enumerate(data_list):
    plt.subplot(2,3,i+1)
    plt.scatter(data_single[:,:-1],data_single[:,-1],s=scalar_list[i],marker=merker_list[i],c=color_list[i])
    plt.title(title_list[i])
plt.suptitle("raw data and standardized dat")
plt.show()
    
```
